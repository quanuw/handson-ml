```python
import numpy as np
from sklearn import datasets
iris = datasets.load_iris()
list(iris.keys())
```

```python
X = iris["data"][:, (2, 3)]
y = iris["target"]
```

```python
# add bias term for every instance X0 =1
X_b = np.c_[np.ones([len(X), 1]), X]
print(X_b)
# random seed so output is reproducible
np.random.seed(2042)
```

```python
# create train, test, validation sets
test_ratio = 0.2
validation_ratio = 0.2
total_size = len(X_b)
test_size = total_size * test_ratio
validation_size = total_size * validation_ratio
train_size = total_size - test_size - validation_size
rnd_indices = np.random.permutation(total_size)
print(rnd_indices)
X_train = X_b[rnd_indices[:int(train_size)]]
y_train = y[rnd_indices[:int(train_size)]]
X_test = X_b[rnd_indices[int(train_size):-int(test_size)]]
y_test = y[rnd_indices[int(train_size):-int(test_size)]]
X_validation = X_b[rnd_indices[-int(test_size)]:]
y_validation = y[rnd_indices[-int(test_size)]:]
```

```python
# turn classes to one hot
def to_one_hot(y):
    n_classes = y.max() + 1
    m = len(y)
    Y_one_hot = np.zeros((m, n_classes))
    Y_one_hot[np.arange(m), y] = 1
    return Y_one_hot
```

```python
Y_train_one_hot = to_one_hot(y_train)
Y_valid_one_hot = to_one_hot(y_validation)
Y_test_one_hot = to_one_hot(y_test)
```

```python
def softmax(logits):
    exps = np.exp(logits)
    exp_sums = np.sum(exps, axis=1, keepdims=True)
    return exps / exp_sums
```

```python
n_inputs = X_train.shape[1] # == 3 (2 features plus the bias term)
n_outputs = len(np.unique(y_train))   # == 3 (3 iris classes)
```

```python
eta = 0.01
n_iterations = 5001
m = len(X_train)
epsilon = 1e-7
alpha = 0.1  # regularization hyperparameter
Theta = np.random.randn(n_inputs, n_outputs)
best_loss = np.infty
for i in range(n_iterations):
    logits = X_train.dot(Theta)
    Y_proba =softmax(logits)
    xentropy_loss = -np.mean(np.sum(Y_train_one_hot * np.log(Y_proba + epsilon), axis=1))
    # Note that the bias term θ0 is not regularized
    l2_loss = 1/2 * np.sum(np.square(Theta[1:])) 
    error = Y_proba - Y_train_one_hot
    loss = xentropy_loss + alpha * l2_loss
    if i % 500 == 0:
        print(i, loss)
        print('error' + str(error))
        print(error.shape)
    gradients = 1/m * X_train.T.dot(error)
    print(gradients.shape)
    Theta = Theta - eta * gradients
    if loss < best_loss:
        best_loss = loss
    else:
        print(i - 1, best_loss)
        print(i, loss, "early stopping!")
        break
```

```python
# predictions for validation set
logits = X_validation.dot(Theta)
Y_proba = softmax(logits)
y_pred = np.argmax(Y_proba, axis=1)
print(y_pred)
accuracy_score = np.mean(y_pred == y_validation)
accuracy_score
```
