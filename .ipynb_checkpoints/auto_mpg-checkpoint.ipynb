```python
import numpy as np
import hashlib
def test_set_check(identifier, test_ratio, hash):
    return hash(np.int64(identifier)).digest()[-1] < 256 * test_ratio

def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
auto_with_id = auto.reset_index()
train_set, test_set = split_train_test_by_id(auto_with_id, 0.2, "index")
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
# print column names
print(auto_num.columns)
# get column index of desired features
acceleration_ix = auto_num.columns.get_loc('acceleration')
weight_ix = auto_num.columns.get_loc('weight')
displacement_ix = auto_num.columns.get_loc('displacement')
cylinders_ix = auto_num.columns.get_loc('cylinders')

# Make a Transformer to add custom attributes
class CombineAttributesAdder(BaseEstimator, TransformerMixin):
    # no *args or **kargs
    def __init__(self, add_cylinders_per_displacement = True):
        self.add_cylinders_per_displacement = add_cylinders_per_displacement
    def fit(self, X, y=None):
        # nothing else to do
        return self
    def transform(self, X, y=None):
        acceleration_per_weight = X[:, acceleration_ix] / X[:, weight_ix]
        if self.add_cylinders_per_displacement:
            cylinders_per_displacement = X[:, cylinders_ix] / X[:, displacement_ix]
            return np.c_[X, acceleration_per_weight, cylinders_per_displacement]
        else:
            return np.c_[X, acceleration_per_weight]
```

```python
attr_adder = CombineAttributesAdder()
auto_extra_attribs = attr_adder.transform(auto.values)
auto_extra_attribs.shape
auto.shape
```
